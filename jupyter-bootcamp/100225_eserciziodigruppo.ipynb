```python
from matplotlib import pyplot as plt

file_path="input.txt"

num1=[]
num2=[]
abs_diff = []

with open(file_path, 'r') as file:
        for line in file:
            parts = line.strip().split()
            if len(parts) == 2:
                n1 = int(parts[0])
                n2 = int(parts[1])
                abs_diff.append(abs(n1-n2))

plt.figure(figsize=(12, 6))
plt.plot(abs_diff, marker='o', linestyle='-', color='#1f77b4', markersize=5, label='|Δ Valore|')
max_index = abs_diff.index(max(abs_diff))
min_index = abs_diff.index(min(abs_diff))
plt.scatter([max_index], [abs_diff[max_index]], color='red', s=100, label='Massimo')
plt.scatter([min_index], [abs_diff[min_index]], color='green', s=100, label='Minimo')
plt.xlabel("Indice della Coppia", fontsize=12)
plt.ylabel("Differenza Assoluta", fontsize=12)
plt.title("Andamento della Differenza Assoluta tra Coppie di Numeri", fontsize=14, fontweight='bold')
```
